### Introduction

HR analytics, also referred to as people analytics, workforce analytics, or talent analytics, involves gathering together, analyzing, and reporting HR data. It enables your organization to measure the impact of a range of HR metrics on overall business performance and make decisions based on data. In other words, HR analytics is a data-driven approach toward Human Resources Management.

Major challenge faced by businesses is employee retention. Managers have to ensure that the new hires fit in with the company culture, get along well with their team and work in a more efficient way with their colleagues. They often fail to meet these requirements, resulting in a high turnover rate. HR analytics helps managers understand the reason behind high turnover rates. They can use this information to take the necessary steps to ensure employee retention.

By predicting the impact of practices, HR analytics give management an opportunity to align strategies with business goals. It enables companies to understand which practices are working and allows them to make changes in accordance with observed analytics. HR professionals can better understand the impact of business initiatives and work on the alignment of action plans to improve efficiencies.

HR analytics helps the HR department collects information, and analyze and benchmark their performance. Organizations can now use data and analytics in order to make better and quick business decisions that previously were very tough and time-consuming processes. They can make smart decisions based on real-time data to enhance the bottom line of their companies.

### Bivariate Analysis

#### 1. What is the attrition rate in all departments ?

```python
df2 = df1.groupby(['Department', 'Attrition'])['Department'].count().reset_index(name='Count')
total_counts = df2.groupby('Department')['Count'].transform('sum')
df2['Percentage'] = (df2['Count'] / total_counts) * 100

plt.figure(figsize=(7, 4))
ax = sns.barplot(x='Department', y='Percentage', hue='Attrition', data=df2, palette=['grey','purple'])

for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 4), textcoords='offset points')

plt.title('Attrition Percentage by Department')
plt.ylabel('Percentage')
plt.show()
```

#### Insights - 

Company is having 3 major departments were Highest attrition can be seen in Sales department by 20.63% followed by Human Resources department with 19.05% and least can be seen in Research and Developement department by 13.84%. It can be said that R&D department is having more stability in it as compared to two other departments.

#### 4. Study worklife balance in different age groups among employees in the company 

```python
bins = [18, 25, 35, 45, 55, 60]
labels = ['18-24', '25-34', '35-44', '45-54', '55-60']
df1['AgeGroup'] = pd.cut(df1['Age'], bins=bins, labels=labels, right=False)

df4 = df1.groupby(['WorkLifeBalance', 'AgeGroup'])[['Age']].count()

df4 = df4.rename(columns={'Age': 'Count'})

df_pivot = df4.pivot_table(index='AgeGroup', columns='WorkLifeBalance', values='Count', fill_value=0).astype(int)

df_pivot
```

```python
plt.figure(figsize=(10,3))
df_pivot.plot(kind='bar', stacked=True, color=['lightblue','pink','grey','purple'])
plt.title('Work Life Balance by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

#### Insights - 

A significant number of employees, especially those within the age brackets of 25-34 and 35-44, consistently expressed satisfaction with their work-life balance, consistently assigning it a rating of 3 out of 4. Across all age groups, the prevalence of a 3 out of 4 rating suggests a generally fair work-life balance.

Interestingly, as age increases, work life balance also increases, this positive trend doesn't hold true here. In the 55-60 age group, ratings of 2 and 3 out of 4 are nearly equal, suggesting that higher-age employees may struggle with managing work-life balance, possibly due to elevated work pressure in senior roles. It's plausible that individuals in this age range are working more to secure their retirement, as settling down for retirement may be a priority.

Conversely, the age group of 18-24 rarely experiences very good work-life balance, possibly indicating that young professionals entering their careers face challenges and significant hustle in their initial phases. Analyzing the data, it becomes apparent that achieving a perfect work-life balance (4 out of 4) is rare across all age groups. The peak of excellent work-life balance is observed in the age groups of 25-34 and 35-44, but it gradually decreases thereafter.
